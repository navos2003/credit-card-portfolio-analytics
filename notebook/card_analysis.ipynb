```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv(r"C:\Users\fazen\OneDrive\Desktop\Kaggle\BankChurners.csv")

df.shape
```

```python
df.head()
```

```python
df.info()
df.describe()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
df = df.loc[:, ~df.columns.str.contains("naive")]
df.columns
df.head()
```

```python
df["util_segment"] = pd.cut(
    df["avg_utilization_ratio"],
    bins=[0, 0.3, 0.7, 1],
    labels=["Low", "Medium", "High"]
)

sns.countplot(x="util_segment", data=df)
plt.title("Utilization Segmentation")
plt.show()
```

```python
df["risk_flag"] = np.where(
    (df["avg_utilization_ratio"] > 0.8) |
    (df["months_inactive_12_mon"] >= 3),
    "high_risk",
    "normal"
)
```

```python
print("Total customers:", len(df))
print("Avg utilization:", df["avg_utilization_ratio"].mean())
print("Total spend:", df["total_trans_amt"].sum())
```

```python
df.groupby("card_category")["total_trans_amt"].mean()
df.groupby("income_category")["avg_utilization_ratio"].mean()
```

```python
sns.histplot(df["avg_utilization_ratio"], bins=30, kde=True)
plt.title("Credit Utilization Distribution")
plt.xlabel("Utilization Ratio")
plt.ylabel("Customer Count")
plt.show()
```

```python
sns.countplot(x="risk_flag", data=df)
plt.title("Risk Distribution")
plt.show()
```

```python
risk_pct = df["risk_flag"].value_counts(normalize=True) * 100
risk_pct.plot(kind="bar")

plt.title("Risk Distribution (%)")
plt.ylabel("Percentage")
plt.show()
```

```python
sns.boxplot(x="risk_flag", y="total_trans_amt", data=df)
plt.title("Transaction Amount by Risk Group")
plt.show()
```
